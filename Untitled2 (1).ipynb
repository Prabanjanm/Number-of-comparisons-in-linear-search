```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

def measure_time(arr, key):
    start = time.time()
    linear_search(arr, key)
    end = time.time()
    return (end - start) * 1000  # convert to milliseconds

def generate_input(n):
    return random.sample(range(1, n * 10), n)

def main():
    print("=== Linear Search Time Analysis for 3 Input Sizes ===")

    # Get three input sizes from user
    while True:
        try:
            sizes = list(map(int, input("Enter 3 input sizes separated by space (e.g., 1000 5000 10000): ").split()))
            if len(sizes) != 3:
                print("❗ Please enter exactly 3 input sizes.")
                continue
            break
        except ValueError:
            print("❗ Invalid input. Enter 3 integers separated by space.")

    times_best = []
    times_avg = []
    times_worst = []

    for n in sizes:
        arr = generate_input(n)

        # Best Case
        key_best = arr[0]
        t_best = measure_time(arr, key_best)

        # Average Case
        key_avg = arr[n // 2]
        t_avg = measure_time(arr, key_avg)

        # Worst Case
        key_worst = -1
        t_worst = measure_time(arr, key_worst)

        times_best.append(t_best)
        times_avg.append(t_avg)
        times_worst.append(t_worst)

        # Print results for each input
        print(f"\n Input size n = {n}")
        print(f"   Best   : {t_best:.4f} ms")
        print(f"   Average: {t_avg:.4f} ms")
        print(f"   Worst  : {t_worst:.4f} ms")

    # Plotting grouped bar chart
    x = np.arange(len(sizes))  # label positions
    width = 0.25

    plt.figure(figsize=(10, 6))
    plt.bar(x - width, times_best, width, label='Best Case', color='green')
    plt.bar(x, times_avg, width, label='Average Case', color='orange')
    plt.bar(x + width, times_worst, width, label='Worst Case', color='red')

    plt.xlabel("Input Size (n)")
    plt.ylabel("Time (ms)")
    plt.title("Linear Search Time Comparison (Best, Average, Worst)")
    plt.xticks(x, sizes)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
